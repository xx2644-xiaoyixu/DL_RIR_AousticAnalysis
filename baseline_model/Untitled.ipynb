```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(len(audio)) / sr

plt.figure(figsize=(10, 4))

if audio.ndim == 1:
    plt.plot(t, audio)
else:
    plt.plot(t, audio[:, 0], label="Left")
    plt.plot(t, audio[:, 1], label="Right", alpha=0.7)

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0.0, 0.3)
plt.title("Waveform")
plt.legend()

plt.show()
```
